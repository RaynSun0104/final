This notebook computes an **Education Index** for each Pittsburgh neighborhood.  
My metric focuses on the "studiousness" or educational attainment potential of a neighborhood. 

**Metric Definition:**  
Education Score = (High School Enrollment + College Undergraduate Enrollment + Graduate/Professional School Enrollment) / Total Population (Age 3+)

The higher the ratio, the better the score (normalized to 0-1).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Setting display options to see all columns once we want to
pd.set_option('display.max_columns', None)
```

## 1. Load Data
I am using the **2015 American Community Survey: School Enrollment** dataset from WPRDC. 

The specific dataset is School Enrollment, All Races.

Link: https://data.wprdc.org/dataset/pittsburgh-american-community-survey-2015-school-enrollment

```python
# Load the dataset
# Read CSV
df = pd.read_csv("school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over.csv")

print(f"Data loaded with shape: {df.shape}")
df.head(3)
```

## 2. Data Cleaning & Feature Extraction

We need to filter out the specific columns relevant to High School, College, and Graduate school enrollment.

**Relevant Columns:**
- `Estimate; Total:` (The denominator)
- `Estimate; Enrolled in school: - Enrolled in grade 9` through `grade 12` (High School)
- `Estimate; Enrolled in school: - Enrolled in college, undergraduate years` (College)
- `Estimate; Enrolled in school: - Graduate or professional school` (Grad School)

```python
# 1. Filter needed columns
needed_cols = [
    'Neighborhood', 
    'Id', 
    'Estimate; Total:',
    'Estimate; Enrolled in school: - Enrolled in grade 9',
    'Estimate; Enrolled in school: - Enrolled in grade 10',
    'Estimate; Enrolled in school: - Enrolled in grade 11',
    'Estimate; Enrolled in school: - Enrolled in grade 12',
    'Estimate; Enrolled in school: - Enrolled in college, undergraduate years',
    'Estimate; Enrolled in school: - Graduate or professional school'
]

edu_df = df[needed_cols].copy()

# 2. Sum up High School students (Grades 9-12)
edu_df['High_School_Total'] = (
    edu_df['Estimate; Enrolled in school: - Enrolled in grade 9'] +
    edu_df['Estimate; Enrolled in school: - Enrolled in grade 10'] +
    edu_df['Estimate; Enrolled in school: - Enrolled in grade 11'] +
    edu_df['Estimate; Enrolled in school: - Enrolled in grade 12']
)

# 3. Rename other columns for easier access
edu_df.rename(columns={
    'Estimate; Enrolled in school: - Enrolled in college, undergraduate years': 'College_Total',
    'Estimate; Enrolled in school: - Graduate or professional school': 'Grad_Total',
    'Estimate; Total:': 'Total_Pop_Age_3plus'
}, inplace=True)

# 4. Calculate the "Advanced Education Population"
# This includes HS, College, and Grad school.
edu_df['Adv_Edu_Pop'] = edu_df['High_School_Total'] + edu_df['College_Total'] + edu_df['Grad_Total']

# Display check
edu_df[['Neighborhood', 'Total_Pop_Age_3plus', 'Adv_Edu_Pop']].head()
```

## 3. Metric Calculation

Now we calculate the ratio:
$$ \text{Education Ratio} = \frac{\text{High School} + \text{College} + \text{Grad School}}{\text{Total Population (3+)}} $$

This ratio represents the density of students pursuing secondary or higher education in the neighborhood.

```python
# Calculate Ratio
# Avoid division by zero just in case
edu_df['Edu_Ratio'] = edu_df['Adv_Edu_Pop'] / edu_df['Total_Pop_Age_3plus']

# Handle potential NaN values if population was 0
edu_df['Edu_Ratio'] = edu_df['Edu_Ratio'].fillna(0)

print("Top 5 neighborhoods by raw Education Ratio:")
print(edu_df.sort_values(by='Edu_Ratio', ascending=False)[['Neighborhood', 'Edu_Ratio']].head(5))
```

## 4. Normalization (Min-Max Scaling)

To combine this with other metrics (Safety and Traffic), normalize this data to a 0-1 scale.
- **0** represents the neighborhood with the lowest education enrollment rate.
- **1** represents the neighborhood with the highest education enrollment rate.

```python
def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())

edu_df['Education_Score'] = min_max_normalize(edu_df['Edu_Ratio'])

# Sort by score to see the best neighborhoods
edu_ranking = edu_df.sort_values(by='Education_Score', ascending=False)

print("Top 10 Neighborhoods by Education Score:")
edu_ranking[['Neighborhood', 'Education_Score']].head(10)
```

## 5. Visualization

Visualizing the top 15 neighborhoods based on the Education Metric.

```python
top_15 = edu_ranking.head(15)

plt.figure(figsize=(10, 6))
# Invert order so the best is at the top
plt.barh(top_15['Neighborhood'][::-1], top_15['Education_Score'][::-1], color='skyblue')
plt.xlabel('Education Score (0-1)')
plt.title('Top 15 Neighborhoods - Education Opportunity')
plt.tight_layout()
plt.show()
```

## 6. Conclusion

Based on the **Education Index**, the best neighborhoods are primarily those with high concentrations of university students, which makes sense given that college and graduate enrollment significantly weights the metric.

**North Oakland** and **Central Oakland** dominate this metric, likely due to the presence of the University of Pittsburgh. **Bluff** (Duquesne University area) also scores very high.

This metric will be combined with Safety and Traffic Quietness to find a balanced "Best" neighborhood that is both safe, quiet, and educated.

Exporting the score to CSV for the group combined notebook.

```python
# Exporting just the columns needed for the group project
output_df = edu_df[['Neighborhood', 'Education_Score']].rename(columns={'Education_Score': 'value'})

# Saving to csv
output_df.to_csv('education_score.csv', index=False)
print("File 'education_score.csv' saved successfully.")
```
